```python
import pymc3 as pm
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
model = pm.Model()
with model:
    x = pm.Normal('x', 0, 1) # prior
    
    trace = pm.sample()
```

```python
pm.summary(trace)
```

```python
model = pm.Model()
with model:
    x = pm.Normal('x', 0, 1) # prior
    obs = pm.Normal('obs', x, 0.01, observed=0.5) # likelihood
    trace = pm.sample()
```

```python
pm.summary(trace)
```

```python
N = 20
RCK = -1.61
RCS = 0.01
K_true = np.random.randn(N) * RCS + RCK
sigma_K_obs = np.ones(N) * 0.05
K_obs = K_true + np.random.randn(N) * sigma_K_obs

ax = sns.distplot(K_true, label=r'$K_{\rm true}$')
ax = sns.distplot(K_obs, label=r'$K_{\rm obs}$')
ax.legend()
```

```python
model = pm.Model()
with model:

    
    age_true_s = pm.Normal('age_true_s', 0, 1, shape=N)
    age_true = pm.Deterministic('age_true', age_true_s * sig + mu)
    
    Kobs = pm.Normal('Kobs', K_true, sigma_K_obs, observed=K_obs)
    
    mu = pm.Normal('mu', 1, 10)
    sig = pm.Lognormal('sig', pm.math.log(0.01), 1)
    
    trace = pm.sample()
```

```python
pm.traceplot(trace)
```

```python
log_period = [1,2,3,4,5]

model = pm.Model()
with model:

    age_true = pm.Deterministic('age_true', pm.Beta('age_true_s', 2, 2, shape=5) * 13.8)
    
    age_obs = pm.Normal('age_obs', age_true, 1.0, observed=np.array([1,2,3,4,5]))
    
    cov_func = pm.gp.cov.ExpQuad(1, ls=3.0)
    gp = pm.gp.Marginal(cov_func=cov_func)
    
    y_ = gp.marginal_likelihood("y", X=age_true[:, None], y=log_period, noise=1.0)

    
    trace = pm.sample()
```

```python
pm.traceplot(trace)
```
