```python
import numpy as np
import matplotlib.pyplot as plt
import pymc3 as pm 
```

```python
x = np.array([1,3,5,7])[:, None]
y_true = np.array([1.3, 2.1, 0.1, -0.7])
y_unc = np.array([0.2, 0.1, 0.1, 0.2])
y = y_true + np.random.randn(len(x))*y_unc
```

```python
plt.errorbar(x, y, yerr=y_unc, linestyle='none')
```

```python
sigma = 1.0 
marginal_gp_model = pm.Model()
with marginal_gp_model:
    l = pm.Normal('l', 3, 0.01)
    
    cov_func = sigma**2 * pm.gp.cov.ExpQuad(1, ls=l)
    
    gp = pm.gp.Marginal(cov_func=cov_func)

    y_ = gp.marginal_likelihood("y", X=x, y=y, noise=y_unc)
    
    trace = pm.sample()
```

```python
Xnew = np.linspace(0, 8, 100)[:, None]

with marginal_gp_model:
    f_star = gp.conditional("f_star", Xnew=Xnew)
    pred_samples = pm.sample_posterior_predictive(trace, vars=[f_star], samples=200)
```

```python
fig, ax = plt.subplots()

from pymc3.gp.util import plot_gp_dist

plt.errorbar(x, y, yerr=y_unc, linestyle='none')
plot_gp_dist(ax, pred_samples["f_star"], Xnew)
```

```python
x_true = np.array([1,3,5,7])
y_true = np.array([1.3, 2.1, 0.1, -0.7])
x_unc = np.array([0.5, 0.5, 0.5, 0.5])
y_unc = np.array([0.2, 0.1, 0.1, 0.2])
y = y_true + np.random.randn(len(x_true))*y_unc
x = x_true + np.random.randn(len(x_true))*x_unc
```

```python
fig, ax = plt.subplots()
ax.errorbar(x, y, xerr=x_unc, yerr=y_unc, linestyle='none')
plot_gp_dist(ax, pred_samples["f_star"], Xnew)
```

```python
sigma = 1.0 
l = 3.0
marginal_gp_model = pm.Model()
with marginal_gp_model:
    x_true = pm.Normal('x_true', 4, 10, shape=4)
    x_obs = pm.Normal('x_obs', x_true, x_unc, observed=x)
    
    cov_func = sigma**2 * pm.gp.cov.ExpQuad(1, ls=l)
    
    gp = pm.gp.Marginal(cov_func=cov_func)

    y_ = gp.marginal_likelihood("y", X=x_true[:, None], y=y, noise=y_unc)
    
    trace = pm.sample(start={'x_true': x})
```

```python
pm.traceplot(trace)
```

```python
Xnew = np.linspace(0, 8, 100)[:, None]

with marginal_gp_model:
    f_star = gp.conditional("f_star", Xnew=Xnew)
    pred_samples = pm.sample_posterior_predictive(trace, vars=[f_star], samples=200)
```

```python
fig, ax = plt.subplots()

from pymc3.gp.util import plot_gp_dist

ax.errorbar(x, y, xerr=x_unc, yerr=y_unc, linestyle='none')
plot_gp_dist(ax, pred_samples["f_star"], Xnew)
```
